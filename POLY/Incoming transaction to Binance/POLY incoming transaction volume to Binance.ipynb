```python
from google.cloud import bigquery
import pandas as pd

client = bigquery.Client()

# In the query we get all transactions to Binance Accounts (Binance has 8 major accounts) from the POLY token
query = """
SELECT 
  *
FROM
  `bigquery-public-data.crypto_ethereum.token_transfers` AS transfers
WHERE to_address IN 
('0x3f5ce5fbfe3e9af3971dd833d26ba9b5c936f0be',
'0xd551234ae421e3bcba99a0da6d736074f22192ff',
'0x564286362092d8e7936f0549571a803b203aaced',
'0x0681d8db095565fe8a346fa0277bffde9c0edbbf',
'0xfe9e8709d3215310075d67e3ed32a380ccf451c8',
'0x4e9ce36e442e55ecd9025b9a6e0d88485d628a67',
'0xbe0eb53f46cd790cd13851d5eff43d12404d33e8',
'0xf977814e90da44bfa03b6295a0616a897441acec')
AND token_address='0x9992ec3cf6a55b00978cddf2b27bc6882d88d1ec'
"""
```

```python
query_job = client.query(query)

iterator = query_job.result(timeout=30)
rows = list(iterator)

# Transform the rows into a nice pandas dataframe
df = pd.DataFrame(data=[list(x.values()) for x in rows], columns=list(rows[0].keys()))

# Look at the first 10
df.head(10)
```

```python
# We convert the values (stored as strings) to floats
df.value=df.value.astype(float)
```

```python
# The 8 major addresses of Binance
binance=['0x3f5ce5fbfe3e9af3971dd833d26ba9b5c936f0be',
'0xd551234ae421e3bcba99a0da6d736074f22192ff',
'0x564286362092d8e7936f0549571a803b203aaced',
'0x0681d8db095565fe8a346fa0277bffde9c0edbbf',
'0xfe9e8709d3215310075d67e3ed32a380ccf451c8',
'0x4e9ce36e442e55ecd9025b9a6e0d88485d628a67',
'0xbe0eb53f46cd790cd13851d5eff43d12404d33e8',
'0xf977814e90da44bfa03b6295a0616a897441acec']

# We select the transactions that are not going to other binance accounts.
external_trans=df.loc[~df['from_address'].isin(binance)]
```

```python
# We get the incoming transactions of poly volume to binance by block_timestamp
res=external_trans.groupby('block_timestamp').sum().sort_values('block_timestamp').value
```

```python
# save the results to csv
res.to_csv('external_trans.csv')
```
