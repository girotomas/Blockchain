```
import requests
import re
import time
import pandas as pd
import json
```

```
def getData(i):
  time.sleep(1)
  headers = {'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_10_1) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/39.0.2171.95 Safari/537.36'}
  r = requests.get('https://etherscan.io/tokens?p={}'.format(i), headers=headers)
  text = r.text
  regex = '''\/token\/(0x[\da-fA-F]+?)'>([^<]+?)<.*?<p[^>]+?>([^<]+?)<\/p.+?<td[^>]+?>([^<]+?)<.*?<td>(\$[\d.,]+?)<.*?(\$[\d.,]+?)(?:&|<).*?<td>([\d.,]+?)<'''
  res =  re.findall(regex, text)
  return res
```

```
res = []
for i in range(1, 20):
  res += getData(i)
df = pd.DataFrame(res, columns = ['address', 'name', 'description', 'price', 'daily_volume', 'market_cap', 'holders'])
df.head()
```
